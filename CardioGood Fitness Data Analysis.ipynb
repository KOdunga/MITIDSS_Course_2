## Q3. Which product is more popular among males or females

```python
pd.crosstab(data['Product'],data['Gender'])
```

```python
sns.countplot(x = 'Gender',hue = 'Product', data = data)
```

## Q4. Is the product purchase affected by the marital status of the customer?
